```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\hp\\OneDrive\\Desktop\\smart loan system.csv")
```

```python
df
```

### Add column- Risk Level

```python
df['Risk_Level'] = np.where(
    (df['Num_Missed_Payments'] > 2) & (df['Collection_Attempts'] > 3),
    'High Risk',
    'Low Risk'
)
```

```python
df
```

### Change data type 

```python

df['Monthly_EMI']=df['Monthly_EMI'].astype('int')
```

```python
df['Monthly_EMI'].dtypes
```

```python
df.columns
```

```python
df.info()
```

### Total Loan Amount

```python
total_loan = df['Loan_Amount'].sum()
total_loan_million = total_loan / 1_000_000

print(f"Total Loan Amount: {total_loan_million:.2f} Million")
```

###  Average Loan Amount

```python
avg_loan = df['Loan_Amount'].mean()
avg_loan_million = avg_loan / 1_000_000

print(f"Average Loan Amount: {avg_loan_million:.2f} Million")
```

### High Risk Borrowers Count

```python
high_risk_count = df[df['Risk_Level'] == 'High Risk'].shape[0]
print("High Risk Borrowers:", high_risk_count)
```

### Average Loan by Risk Level

```python
avg_loan_risk = df.groupby('Risk_Level')['Loan_Amount'].mean() / 1_000

print(avg_loan_risk.round(1).astype(str) + " K")
```

### Gender wise loan Applications

```python
gender_count = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    gender_count.values,
    labels=gender_count.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#800080', '#ff7f0e']
)

plt.title("Gender-wise Loan Applications")
plt.show()

```

### Average loan amount by loan type

```python
avg_loan_type = df.groupby('Loan_Type')['Loan_Amount'].mean() / 1_000
avg_loan_type = avg_loan_type.round(0).astype(int)

plt.figure(figsize=(8,5))

plt.gcf().set_facecolor('#f5f5f5')
plt.gca().set_facecolor('#eef7fa')

plt.barh(
    avg_loan_type.index,
    avg_loan_type.values,
    color='#000080'
)

plt.title("Average Loan Amount by Loan Type (in K)", fontweight='bold')
plt.xlabel("Average Loan Amount (K)")
plt.ylabel("Loan Type")

# 🔹 ADD LABELS
for i, v in enumerate(avg_loan_type.values):
    plt.text(v -100, i, f"{v}K", va='center',color='white', fontweight='bold')

plt.grid(False)
plt.show()
```

###  Identify high-risk borrowers 

```python
risk_count = df['Risk_Level'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    risk_count.values,
    labels=risk_count.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Borrower Risk Level Distribution")
plt.show()
```

### What is the distribution of payment history

```python

payment_count = df['Payment_History'].value_counts()

plt.figure()
plt.bar(payment_count.index, payment_count.values, color=['#008080','#000080','#800000'])
plt.title("Payment History Distribution")
plt.xlabel("Payment History"),
plt.ylabel("Count")
plt.show()
```

### Calculate the Average loan amount by number of dependence

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Group data
dep_loan = df.groupby('Num_Dependents')['Loan_Amount'].mean()

plt.figure(figsize=(6,4))
#  Background colors
plt.gcf().set_facecolor('#f2f4f7')      # outer background
plt.gca().set_facecolor('#e8f0f2')      # chart area background


# Line chart
plt.plot(
    dep_loan.index,
    dep_loan.values,
    marker='o',
    color='purple',
    linewidth=2
)

# Titles & labels
plt.title("Average Loan Amount by Number of Dependents", fontweight='bold')
plt.xlabel("Number of Dependents")
plt.ylabel("Average Loan Amount (K)")

#  Y-axis in K
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f'{int(x/1000)}K')
)

# Add data labels (in K)
for x, y in zip(dep_loan.index, dep_loan.values):
    plt.text(
        x,
        y,
        f"{int(y/1000)}K",
        ha='center',
        va='bottom',
        fontweight='bold'
    )

# ✅ Remove grid
plt.grid(False)

plt.show()
```

### Loan type wise average EMI

```python
avg_emi_type = df.groupby('Loan_Type')['Monthly_EMI'].mean().round(0).astype(int)

plt.figure(figsize=(7,4))

# 🔹 Background colors
plt.gcf().set_facecolor('#f5f5f5')   # full figure background
plt.gca().set_facecolor('#eef7fa')   # chart background

bars = plt.bar(
    avg_emi_type.index,
    avg_emi_type.values,
    color='#800080'
)

plt.title("Average EMI by Loan Type", fontweight='bold')
plt.xlabel("Loan Type")
plt.ylabel("Average EMI")

# 🔹 Add labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height}",
        ha='center',
        va='bottom',
        fontweight='bold'
    )

plt.grid(False)
plt.show()
```

## Key Insights

#### 1. Credit history is the most critical factor influencing loan approval decisions.

#### 2. Applicants with higher and stable income have better chances of loan approval.

#### 3. Higher loan amounts reduce approval probability, as they increase financial risk.

#### 4. Applications with a co-applicant show higher approval rates due to shared repayment responsibility.

#### 5. Longer loan terms are more likely to be approved as they lower monthly repayment burden.

#### 6. Self-employed applicants face slightly lower approval rates compared to salaried applicants.
